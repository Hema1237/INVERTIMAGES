```python
import cv2
import os

# Load the image
image = cv2.imread(r"C:\Users\HEMA M\OneDrive\Pictures\abc-text.jpg")

# Create a directory to save the inverted images
output_dir = 'inverted_images'
os.makedirs(output_dir, exist_ok=True)

# Generate and save 10 inverted images
for i in range(10):
    inverted_image = cv2.bitwise_not(image)  # Invert the image
    cv2.imwrite(os.path.join(output_dir, f'inverted_image_{i+1}.jpg'), inverted_image)

print("10 inverted images saved successfully.")
```

```python
import cv2
import os
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r"C:\Users\HEMA M\OneDrive\Pictures\abc-text.jpg")

# Create a directory to save the inverted images
output_dir = 'inverted_images'
os.makedirs(output_dir, exist_ok=True)

# List to hold inverted images for display
inverted_images = []

# Generate and save 10 inverted images
for i in range(10):
    inverted_image = cv2.bitwise_not(image)  # Invert the image
    inverted_images.append(inverted_image)  # Store for display
    cv2.imwrite(os.path.join(output_dir, f'inverted_image_{i+1}.jpg'), inverted_image)

# Display the inverted images
plt.figure(figsize=(15, 10))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(cv2.cvtColor(inverted_images[i], cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
    plt.title(f'Inverted Image {i+1}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```
